# 1) Break, Continue, Pass

```python
# Break Statement

for i in range(1,11):
    if i==6:
        break
    print(i)
```

```python
# Continue Statement

for i in range(1,11):
    if i % 2 == 0:
        continue # Skip Number
    print(i)
```

```python
# Pass Statement

for i in range(5):
    if i==2:
        pass # Do Nothing
    print(i)
```

# 11) isalnum, isalpha, isdigit

```python
# isalnum() - All Characters are alphanumeric (a-z,A-Z,0-9)
# isalpha() - All Characters are alphabets (a-z, A-Z)
# isdigit() - All characters are digits (0-9)
```

```python
# isalnum()

print("Python3".isalnum())
print("Python 3".isalnum())
print("12345".isalnum())
```

```python
# isalpha()

print("Python".isalpha())
print("Python3".isalpha())
print("Hello World".isalpha())
```

```python
# isdigit()

print("12345".isdigit())
print("123.45".isdigit())
print("12a34".isdigit())
```
