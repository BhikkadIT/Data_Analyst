# 1) Class Relationship : Class relationship define how classes interact with each other

## 1) Aggregation: a relationship where one class contains another class as a member, but both can exist independently.

```python
# real world example : Car has Engine (Engine can exist without Car)
```

```python
class Engine:

    def __init__(self,horsepower):
        self.horsepower = horsepower

    def start(self):
        print(f"Engine with {self.horsepower} HP Strted")
```

```python
class Car:

    def __init__(self,brand,engine):
        self.brand = brand
        self.engine = engine

    def start_car(self):
        print(f"{self.brand} starting...")
        self.engine.start()
```

```python
engine1 = Engine(150)
```

```python
car1 = Car("Maruti",engine1)
car1.start_car()
```

# 4) Constructor in Inheritance

## Ex- 1: Accessing Parent Private Attribute with Getter Method

```python
class Parent:
    def __init__(self):
        self.__private_data = "Secure data"

    # Getter Method
    def get_private_data(self):
        return self.__private_data

class Child(Parent):
    def __init__(self):
        super().__init__()  # Super Keyword (super Method) - Call Parent Constructor

    def display(self):
        print(f"Data:{self.get_private_data()}")
        
child = Child()
child.display()
```

## Ex-2 : if child Has Constructor , Parent Constructor NOT Called.

```python
class Parent:
    def __init__(self):
        self.parent_data = "Parent Data"
        print("Parent constructor called")

class Child(Parent):
    def __init__(self):
        self.child_data = "Child Data"
        print("Child Constructor Called")

child = Child()
```

```python
print(child.parent_data)
```

```python
class Parent:
    def __init__(self):
        self.parent_data = "Parent Data"
        print("Parent constructor called")

class Child(Parent):
    def __init__(self):
        super().__init__()  # Super keyword 
        self.child_data = "Child Data"
        print("Child Constructor Called")

child = Child()
```

```python
print(child.parent_data)
```

## Ex-3: If child Don't have Constructor, Parent Constructor is Called

```python
class Parent:
    def __init__(self):
        self.parent_data = "Parent Data"
        print("Parent constructor called")

class Child(Parent):
    def display(self):
        print(f"Data:{self.parent_data}")

child = Child()
```

# 5) Method Overriding : child class provide its own implementation of a method that exists in parent class

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog Barks")

class Cat(Animal):
    def sound(self):
        print("Cat Meows")

dog = Dog()
dog.sound()
cat = Cat()
cat.sound()
```

```python
# 6) Super Keyword : super() is used to call methods from parent class

# Use: 1)Call Parent Constructor
#      2)Call Parent Methods
#      3)Access Parent Attribute
```

```python
class Parent:
    def show(self):
        print("Parent show Method")

class Child(Parent):
    def show(self):
        super().show() # Call Parent Method
        print("Child Show Method")

child = Child()
child.show()
```

```python
class Grandparent:
    def display(self):
        print("Grandparent")

class Parent(Grandparent):
    def display(self):
        super().display()
        print("Parent")

class Child(Parent):
    def display(self):
        super().display()
        print("Child")

child = Child()
child.display()
```

# 11) Super Constructor

```python
class Vehicle:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

class Car(Vehicle):
    def __init__(self,brand,model,fuel_type):
        super().__init__(brand,model)
        self.fuel_type = fuel_type

car = Car("Maruti","Swift","Petrol")
print(car.brand)
print(car.model)
print(car.fuel_type)
```

# 12) Types of Inheritance

## 1) Single Inheritance : One Child class inherits from one Parent class

```python
class Parent:
    def parent_method(self):
        print("Parent Method")

class Child(Parent):
    def child_method(self):
        print("Child Method")

child = Child()
child.parent_method()
child.child_method()
```

## 2) Multilevel Inheritance - Child Become parent for next class

```python
class Grandparent:
    def gp_method(self):
        print("Grandparent Method")

class Parent(Grandparent):
    def p_method(self):
        print("Parent method")

class Child(Parent):
    def c_method(self):
        print("Child Method")

child = Child()
child.gp_method()
child.p_method()
child.c_method()
```

# 3) Hierarchical Inheritance : Multiple Child classes inherit from one Parent Class

```python
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    def barks(self):
        print("Barking")

class Cat(Animal):
    def meow(self):
        print("Meowing")

class Cow(Animal):
    def moo(self):
        print("Mooing")

dog = Dog()
dog.eat()
dog.barks()
```

## 4)Multiple Inheritance : One child Class inherits from Multiple Parent Classes

```python
class Father:
    def father_method(self):
        print("Father Method")

class Mother:
    def mother_method(self):
        print("Mother Method")

class Child(Father, Mother):
    def child_method(self):
        print("Child Method")

child = Child()
child.father_method()
child.mother_method()
child.child_method()
```

# 5)Hybrid Inheritance : Combination of two or more Types of Inheritance

```python
class A:
    def a_method(self):
        print("Method A")

class B(A):
    def b_method(self):
        print("Method B")

class C(A):
    def c_method(self):
        print("Method C")

class D(B,C):
    def d_method(self):
        print("Method D")

d = D()
d.a_method()
d.b_method()
d.c_method()
d.d_method()
```

# 13) Polymorphism : 

# Many Forms - Same Interface with Different Implementations.

## 1)Method Overriding - Same Method name in parent and Child with Different Implementation.

```python
class Shape:
    def area(self):
        print("Calculate Area")

class Rectangle(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius**2

rect = Rectangle(10,5)
print(rect.area())

circle = Circle(7)
print(circle.area())
```

## 2) Method Overloading : python doesn't support method overloading, but we can achieve it using Default Argument.

```python
class Shape:
    def area(self,radius):
        return 3.14*radius*radius

    def area(self,a,b):
        return a*b

s = Shape()
s.area(2)
s.area(3,4)
```

```python
class Shape:
    def area(self,a,b=0):  # b = Default Argument
        if b ==0:
            return 3.14*a*a
        else:
            return a*b

s = Shape()
print(s.area(2))
print(s.area(3,4))
```
