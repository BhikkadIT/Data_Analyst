# Adding Items to a list

```python
# Methods
# 1) Append
# 2) Extend
# 3) insert
```

```python
# 1) Append - Add Single Items at the end

fruits = ["Apple","Banana"]

fruits.append("Mango")
print(fruits)
```

```python
fruits.append("Grapes")
print(fruits)
```

```python
# 2)Extend - Add Multiple Items at end

list1 = [1,2,3]
list2 = [4,5,6]

list1.extend(list2)
print(list1)
```

```python
list2.extend(list1)
print(list2)
```

```python
# 3)insert - Add items at specific Position

fruits = ["Apple","Mango","Orange"]
fruits.insert(1,"Banana")
print(fruits)
```

```python
fruits.insert(0,"Grapes")
print(fruits)
```

```python
# Editing items in a list

fruits = ["Apple","Banana","Mango"]

fruits[1] = "Grapes"
print(fruits)
```

```python
numbers = [1,2,3,4,5]
numbers[1:4] = [20,30,40]
print(numbers)
```

```python
numbers[2:15] = [14,15,16,17,18,19]
print(numbers)
```

# Deleting items in a list

```python
# remove - Remove First Occurance of Value

fruits = ["Apple","Banana","Mango","Banana"]
fruits.remove("Banana")
print(fruits)
```

```python
name = ["Ajay","Vijay","Digvijay","Parajay","Vijay","Vijay"]
name.remove("Vijay")
print(name)
```

```python
# pop

fruits = ['Apple','Banana','Mango']
remove = fruits.pop()
print(remove)
print(fruits)
```

```python
removed = fruits.pop(0)
print(removed)
print(fruits)
```

```python
# Clear
name = ["Sachin","Hardik","Virendra","Gautam"]

name.clear()
print(name)
```

# List Comprehension

```python
# list comprehension : way to create a list using a single line of code
```

```python
# syntax : [Expression for item in iterable if condition ] 
```

```python
squares = []
for x in range(1,6):
    squares.append(x**2)
print(squares)
```

```python
squaures = [ x**2 for x in range(1,6)]
print(squares)
```

```python
# Even Numbers only

even = [ x for x in range(1,11) if x % 2 == 0]
print(even)
```

# Zip Function : Combines multiple iterables into tuples

```python
names = ["Priya", "Rahul","Amit"]
ages = [22,25,28]

combined = list(zip(names,ages))
print(combined)
```

```python
names,ages = zip(*combined)
print(list(names))
print(list(ages))
```

```python
print(names)
print(ages)
```
