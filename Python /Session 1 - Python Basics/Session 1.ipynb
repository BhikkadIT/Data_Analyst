# 7. Type Conversion / Type Casting

## a) implicit Conversion

```python
a = 10
b = 3.5
c = a + b
```

```python
print(a, type(a))
print(b, type(b))
print(c, type(c))
```

## b) Explicit Conversion

```python
# string to Integer

x = "100"
y = int(x)
print(y+50)
```

```python
print(x, type(x))
print(y,type(y))
```

```python
x = "150"
y = 50
print(x+y)
```

```python
# integer to Float

a = 25
b = float(a)
print(b)
```

```python
# float to integer
c = 3.05
d = int(c)
print(d)
```

```python
# Number to string
age = 25
age_str = str(age)
print("Age :" + age_str)
```

```python
age = 35
age_str = "35"
print(age + age_str)
```
