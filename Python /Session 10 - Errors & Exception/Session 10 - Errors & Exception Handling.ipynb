# Try - Except - Else

```python
try:
    num1 = int(input("Enter Number 1:"))
    num2 = int(input("Enter Number 2:"))
    result = num1/num2
except ValueError:
    print("Enter Valid Number")
except ZeroDivisionError:
    print("Cannot Divide by Zero")
else:
    print(f"Division Successfull: {result}")
```

# Try - Except - Else - Finally

```python
def divide_Number():
    try:
        num1 = int(input("Enter number 1:"))
        num2 = int(input("Enter number 2:"))
        result = num1/num2
    except ValueError:
        print("Enter valid Integer")
    except ZeroDivisionError:
        print("abe gadhe zero se kabhi divide hota hai kya?")
    else:
        print(f"Result:{result}")
        return result
    finally:
        print("Function Execution Completed")
```

```python
divide_Number()
```

```python
def check_voting_age(age):
    if age<0:
        raise ValueError("Age can not be negative")
    elif age<18:
        raise ValueError("Age must be 18 or above")
    else:
        print(f"Age {age} is valid")
```

```python
check_voting_age(15)
```

```python
try:
    check_voting_age(-10)
except ValueError as e:
    print(e)
```

```python
def withdraw_Money(balance,amount):
    if amount > balance:
        raise ValueError("Paise nahi hai tere pass")
    return balance - amount
```

```python
remaining = withdraw_Money(1000,1500)
print(remaining)
```

```python
try:
    remaining = withdraw_Money(1000,1500)
except ValueError as e:
    print(e)
```

# Creating Custom Exception

```python
class InvalidAgeError(Exception):
    pass

def check_voting_age(age):
    if age<18:
        raise InvalidAgeError("You must be 18 or older to vote")
    else:
        print("You are eligible to vote")
```

```python
try:
    check_voting_age(8)
except InvalidAgeError as e:
    print(e)
```
