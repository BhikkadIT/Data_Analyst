# Encapsulation :

# Encapsulation means hiding internal details and restricting direct access of data

# a) Public Attrtibute

```python
class Student:

    def __init__(self,name,age):
        self.name = name  # public attribute
        self.age = age    # public attribute
        
```

```python
student1 = Student("vijay",20)
```

```python
print(student1.age)
```

```python
print(student1.name)
```

```python
student1.age = "hehehe"
```

# b) protected attribute

```python
class Student:

    def __init__(self,name,age):
        self.name = name  # public attribute
        self._age = age    # protected attribute
        
```

```python
student2 = Student("vijay",45)
```

```python
print(student2.age)
```

```python
print(student2._age)
```

```python
student2._age = "hehehe"
```

# d) Getter & Setter Method

```python
class Student:

    def __init__(self,name,age):
        self.name = name
        self.__age = age  # private attribute

    def get_age(self):  # getter method (value ko get karne ke liye)
        return self.__age

    def set_age(self,age):  # setter method(value ko set karne ke liye)
        if type(age)==int:
            if age>0:
                self.__age = age
```

```python
student4 = Student("Akshay",49)
print(student4.__age)
```

```python
print(student4.get_age())
```

```python
print(student4.set_age(23))
```

```python
print(student4.get_age())
```

```python
student4.set_age("hehehe")
```

```python
print(student4.get_age())
```

# 7. Collection of objects

## you can store multiple objects in lists, dictionary, or other collection

```python
class Student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def display(self): #  instance method (object ke method)
        print(f"Name: {self.name}, Marks:{self.marks}")
```

```python
S1 = Student("Akshay",85)
S2 = Student("Ajay",55)
S3 = Student("Sanjay",90)
```

```python
students = [S1,S2,S3]
```

```python
for student in students:
    student.display()
```

# 9) Static Method :

# static method belong to the class, not to any specific object

```python
class Calculator:

    @staticmethod # Decorator
    def add(a,b):
        return a+b

    @staticmethod
    def multiply(a,b):
        return a*b
```

```python
print(Calculator.add(5,10))
print(Calculator.multiply(5,10))
```

```python
calc = Calculator()
```

```python
print(calc.add(15,25))
```
