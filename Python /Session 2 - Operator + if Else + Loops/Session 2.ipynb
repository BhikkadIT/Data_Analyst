```python
number = int(input("Enter a number :"))
if number % 2 == 0:
             print("The Number is Even")
else:
    print("The number is Odd")             
```

```python
a = int(input("Enter First Number"))
b = int(input("Enter Second Number"))

if a > b:
    print("The First Number is greater")
elif a<b:
    print("The Second Number is greater")
else:
    print("Both Numbers are equal")
```

```python
# IF-ELIF-ELSE Statement

marks = int(input("Enter your marks"))
if marks >= 90:
    print("Grade A")
elif marks>=75:
    print("Grade B")
elif marks >=60:
    print("Grade C")
elif marks >=40:
    print("Grade D")
else:
    print("Grade E (Fail)")
```

# 4. Loops In Python

```python
## A) While Loop
# syntax 
# while condition:
    # code to execute
    # update condition
```

```python
# Basic Example :

count = 1

while count<=5:
    print(count)
    count += 1   # count = count + 1
```

```python
num = 5
i = 1

while i<=10:
    print("Num",num * i)
    i += 1  # i = i + 1
```

```python
# Break Statement

count = 1

while True:
    print(count)
    count +=1
    if count > 5:
        break
```

```python
# Continue Statement

num = 0
while num<10:
    num +=1
    if num%2 ==0:
        continue # Skip Number
    print(num)    
```

```python
count = 1

while count<=5:
    print(count)
    count += 1
else:
    print("Loop Completed Succesfully")
```

```python
# B) For LOOP : iterate over sequence

# syntax :
#      for variable in sequence
#      code to execute
```

```python
fruits = ["Apple","Banana","Mango"]

for fruit in fruits:
    print(fruit)
```

```python
name = "Python"
for Akshay in name:
    print(Akshay)
```

```python
for i in range(1,8):  # (start, end)
    print(i)
```

```python
for i in range(2,11,3):  # (Start, end, step)
    print(i)
```

```python
# Break

for i in range(1,11):
    if i==6:
        break
    print(i)    
```

```python
# Continue 

for i in range(1,11):
    if i%2 ==0:
        continue # Skip Number
    print(i)
```

```python
# C) Nested Loop

# Basic Nested Loop

for i in range(1,4):
    for j in range(1,4):
        print("i",i,"j",j)
```

```python
# Start Pattern
for i in range(1,6):
    for j in range(i):
        print("*",end="")
    print()
```
