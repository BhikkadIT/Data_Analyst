# 2)Scope & LEGB Rule

- Scope : Scope is defines the region in code where a variable is accessible

# LEGB Rule : 
- Local Scope - (L) - inside current Function
- Enclosing Scope - (E) - Inside Enclosing Function
- Gloab Scope - (G) - At Module Level
- Builtin Scope - (B) - python Builtin Scope

```python
x = 4   # Global Scope (G)  

def outer():
    x = 10  # Enclosing Scope (E)

    def inner():
        x = 7   # Local Scope (L)
        print("Inner:",x)

    inner()
    print("Outer:",x)

outer()
print("Global:",x)
```

```python
# Local Scope :

def local_function():
    local_var = 'i am local' # Local Scope (Local Variable)
    print(local_var)

local_function()
```

```python
print(local_var)
```

```python
# Global Scope :

Global_var = ' i am global'

def global_function():
    print(Global_var)

global_function()
print(Global_var)
```

```python
counter = 0

def increment():
    global counter
    counter +=1  # counter = Counter + 1
    print("Counter",counter)

increment()
increment()
print(counter)
```

# Decorators :

- A Decorator is a function that modifies the behaviour of another function without changing its code.

```python
def my_decorator(func):
    def wrapper():
        print("**********************")
        func()
        print("**********************")
    return wrapper

def say_hello():
    print("Hello")

hello = my_decorator(say_hello)
hello()
```

```python
def my_decorator(func):
    def wrapper():
        print("**********************")
        func()
        print("**********************")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```

# Decorator with Function Argument:

```python
def decorator(func):
    def wrapper(*args):
        print("Befor Function call")
        result = func(*args)
        print("After Function call")
        return result
    return wrapper

@decorator
def add(a,b):
    print(f" Adding {a} + {b}")
    return a+b

result = add(5,10)
print("Result",result)
```

```python
# Practical Example : Login Required Decorator

def login_required(func):
    def wrapper(user):
        if user.get("Logged in"):
            return func(user)
        else:
            print("Please login first")
    return wrapper
        
@login_required
def view_profile(user):
    print(f" Welcome {user['name']}")

user1 = {"name":"Akshay","Logged in":True}
user2 = {"name":"Rahul","Logged in":False}

view_profile(user1)
view_profile(user2)
```

```python
def uppercase_decorator(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

def exclamation_decorator(func):
    def wrapper():
        result = func()
        return result + "!!!"
    return wrapper

@uppercase_decorator
@exclamation_decorator
def greet():
    return "hello"

print(greet())
```
