# Class with Method

```python
class Student:
    def __init__(self,name,marks): 
        self.name = name # Attribute
        self.marks = marks # Attribute

    def display_info(self):
        print(self.name)
        print(self.marks)
```

```python
S1 = Student("Rahul",85)      #  Syntax: Objectname = Classname()
```

```python
print(S1.name)
```

```python
print(S1.marks)
```

```python
S1.display_info()
```

# Constructor with parameter

```python
class Student:
    def __init__(self,name,age,marks):
        self.name = name
        self.marks = marks
        self.age = age
        print(f"Student {name} created")
```

# 5) Method Vs Function :

```python
class Student:
    def __init__(self,name,marks):    # Magic Method 
        self.name = name # Attribute
        self.marks = marks # Attribute

    def display_info(self):  # Method
        print(self.name)
        print(self.marks)
```

```python
S1 = Student("Akshay",85)
S1.display_info()
print(S1)
```

```python
# Function

def power(a,b):
    return a**b
```

# 6) Magic Method : A special method with double underscore before and after name. python calls them automatically

```python
# __name__()

# __str__()
```

```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    # String represenation
    def __str__(self):  # Magic Method
        return f"Student: {self.name} , Marks:{self.marks}"
```

```python
student1 = Student("Rahul",85)
```

```python
print(student1)
```

# 7) Concept of Self: self represents the current instance of the class. it is used to access attributes and methods

```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        print(id(self))

    def display(self):
        print(self.name)
        print(self.age)
```

```python
student1 = Student("Rahul",25)
student1.display()
```

```python
print(id(student1))
```

```python
student1 = Student("Rahul",25)
student1.display()
```

```python
student2 = Student("Vijay",28)
```

```python
print(id(student2))
```
