# File Handling

### Some Theory

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files

### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

# Writing Many Files

```python
files_data = {
    "file1.txt":"Content of file 1\n",
    "file2.txt":"Content of file 2\n",
    "file3.txt":"Content of file 3\n",
}
```

```python
for filename,content in files_data.items():
    file = open(filename,'w')
    file.write(content)
    file.close()
```

# Reading from readline

# readline() read file into single line (line by line)

```python
file = open('data2.txt','w')
file.write("This is the First line of program\n")
file.write("This is the Second line of Program\n")
file.write("This is the Third line of Program\n")
file.close()
```

```python
file = open("data2.txt",'r')
print(file.readline(),end='')
print(file.readline(),end='')
print(file.readline(),end='')
file.close()
```

# Reading Big File Line By Line

```python
file = open('data3.txt','w')
file.write("Python Program")
file.write("Java Program")
file.write("PHP Program")
file.write("C++ program")
file.close()
```

```python
file = open("data3.txt",'r')

while True:
    line = file.readline()
    if not line:
        break
    print(line)
file.close()
```

# Using Context Manager With 

# with Statement automatically opens and close file

```python
file = open('data4.txt','w')
file.write("My Name is Akshay")
file.close()
```

```python
file.write("My Name is Vijay")
```

```python
with open('data4.txt','r') as f:
    content = f.read()
    print(content)
```

```python
f.write("My name is Sachin")
```

# Reading Big File in Chunks

```python
Big_L = ['Hello World' for i in range(1000)]
print(Big_L)
```

```python
with open('big.txt','w') as f:
    f.writelines(Big_L)
```

```python
with open('big.txt','r') as f:
    chunk_size = 10
    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size))
        #f.read(chunk_size)
```

### Problems with working in text mode

- can't work with binary files like images
- not good for other data types like int/float/list/tuples

```python
with open('Gmail_Logo.png','r') as f:
    print(f.read())
```

```python
with open('Gmail_Logo.png','rb') as f:
    print(f.read())
```

```python
with open("Gmail_Logo.png",'rb') as f:
    with open('Gmail_copy.png','wb') as wf:
        wf.write(f.read())
```

# Serialization & Deserialization :

# Serialization : Convert pyton Object into json Format
# Deserializaton : Convert json format into python Object

```python
# Serialization :

import json

student = {
    'name':"Akshay",
    'age':25,
    'grade': [85,90,88]
}
```

```python
with open('student.json','w') as file:
    json.dump(student,file)
```

```python
# Deserialization :

import json

with open('student.json','r') as file:
    student = json.load(file)
```

```python
print(student)
```

```python
print(type(student))
```
