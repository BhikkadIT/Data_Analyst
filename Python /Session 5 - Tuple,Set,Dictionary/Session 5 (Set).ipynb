# 5) Set Opeartions

```python
# Union()
# Intersection()
# Difference()
# Symmetric Difference()
```

```python
# Union
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 | set2 # ( | - union)
print(result)
```

```python
result2 = set1.union(set2)
print(result2)
```

```python
set3 = {7,8,9}
result = set1 | set2 | set3
print(result)
```

```python
# Intersection
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 & set2 # ( & - intersection)
print(result)
```

```python
result = set1.intersection(set2)
print(result)
```

```python
# Difference 
set1 = {1,2,3,4}
set2 = {8,9,5,6}
result = set2 - set1
print(result)
```

```python
result = set1.difference(set2)
print(result)
```

```python
# Symmetric Difference
set1 = {1,2,3,4}
set2 = {3,4,5,6}
result = set1 ^ set2 # ( ^ - Symmetric difference)
print(result)
```

```python
result = set1.symmetric_difference(set2)
print(result)
```

# 8) Set Comprehension

```python
# Way to create a sets using a single line of code
```

```python
# Traditional Way
squares = set()
for x in range(1,6):
    squares.add(x**2)
print(squares)
```

```python
squares = {x**2 for x in range(1,6)}
print(squares)
```

```python
# print even numbers only
evens = {x for x in range(1,11) if x % 2 ==0}
print(evens)
```
