# 6) Operations on a tuple

```python
# Concatenation 
tuple1 = (1,2,3)
tuple2 = (4,5,6)
add = tuple1 + tuple2
print(add)
```

```python
# Repetition
numbers = (1,2)
result = numbers * 3
print(result)
```

```python
# Comparision Operators

tuple1 = (1,2,3)
tuple2 = (1,2,3)
tuple3 = (1,2,4)
```

```python
print(tuple1 == tuple2)
print(tuple1 != tuple3)
print(tuple1 < tuple3)
```

```python
# Membership Operator

fruits = ("Apple", "Banana", "Potato")
print("Potato" in fruits)
print("Potato" not in fruits)
```

```python
print("" in fruits)
```

```python
fruits = ("Apple", "Banana", "Potato")

for fruit in fruits:
    print(fruit)
```

```python
# 8) Difference between list and Tuple

# 1) Mutability - List are Mutable

my_list = [1,2,3]
my_list[0] = 100
print(my_list)
```

```python
my_tuple = (1,2,3) # Tuples are Immuatble
my_tuple[0] = 100
```

```python
# 2) Performance 
# List - Slower
# Tuple - Fast

import timeit
list_time = timeit.timeit("[1,2,3,4,5]",number =1000000)
tuple_time = timeit.timeit("(1,2,3,4,5)",number =1000000)

print("List Creation Time", list_time)
print("Tuple Creation Time", tuple_time)
```

```python
# 3) Memory Camparision

import sys
my_list = [1,2,3,4,5] # required more Memory
my_tuple = (1,2,3,4,5) # required less memory

print("List size", sys.getsizeof(my_list))
print("Tuple size", sys.getsizeof(my_tuple))
```

```python
# Methods Available

my_list = [1,2,3,4,5]
my_tuple = (1,2,3,4,5)

print(dir(my_list))
print(dir(my_tuple))
```

# 9) Tuple Unpacking : - Assigning Tuple items to multiple Variables in one statement

```python
coordinates = (10,20,30)
x, y,z = coordinates

print(x,y,z)
```

```python
name, age, city = ("Rahul", 25,"Nashik")
print(name)
print(age)
print(city)
```

```python
numbers = (1,2,3,4,5)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
a, b , *rest = numbers
print(a)
print(b)
print(rest)
```

# 10) Zip Function :- Combines Multiple iterable into tuple

```python
names = ("Rahul", "Priya", "Amit")
ages = (25,22,28)

combined = tuple(zip(names,ages))
print(combined)
```

```python
# Unzip 

combined = (('Rahul', 25), ('Priya', 22), ('Amit', 28))
names ,ages = zip(*combined)
print(names)
print(ages)
```

```python
tuple1 = (1,2,3,4)
tuple2 = ("a","b")
result = tuple(zip(tuple1, tuple2))
print(result)
```
